```python
import numpy as np
import pandas as pd
import seaborn as sns
import datetime

order_data = pd.read_excel("آنالیز داده.xlsx", usecols =["نوع حواله", "بخش مصرف کننده", "درخواست کننده", "تاريخ ايجاد حواله", 
                                                    "وضعيت كلي حواله", "شرح آیتم", "مقدار اوليه درخواستي", "مقدار تاييد شده", "مقدار تحويل شده", 
                                                    "موجودي كل انبارها", "میانگین مصرف بخش", "مصرف جاری", "میانگین مصرف کل", 
                                                    "حواله در جریان", "حواله بین راهی", "حواله خرید", "مقدار درخواست خريد فعال", "مقدار سفارش خريد فعال",
                                                    "رسيد هاي تحويل نهايي نشده", "خريد تاييد شده ولي تحويل نهايي نشده", 
                                                    "تاریخ دريافت حواله توسط پلنر", "پلنر", "عمليات انجام شده توسط پلنر", "نوع حواله", "وضعيت رديف حواله"])
print("تعداد سطر و ستون:", order_data.shape)
print("لیست ستون ها:\t", order_data.columns.tolist())
#print(order_data.head(10))

print("داده های ناقص/گم شده در هر ستون:\n", order_data.isnull().sum())
print("\n داده های تکراری:", order_data.duplicated().sum())

key_columns = ["مقدار اوليه درخواستي", "مقدار تاييد شده", "مقدار تحويل شده", "موجودي كل انبارها", "میانگین مصرف بخش", "مصرف جاری", "میانگین مصرف کل"]
print("آمار توصیفی ستون های کلیدی:\n", order_data[key_columns].describe())

print("تعداد هر وضعیت کلی هر حواله: \n")
print(order_data["وضعيت كلي حواله"].value_counts())

print("تعداد عملیات انجام شده توسط پلنر ها: \n")
print(order_data["عمليات انجام شده توسط پلنر"].value_counts())

order_data[["نوع حواله", "بخش مصرف کننده"]] = order_data[["بخش مصرف کننده", "نوع حواله"]].astype("category")
order_data["مقدار تحويل شده"] = pd.to_numeric(order_data["مقدار تحويل شده"], errors = "coerce")

#print(order_data.info())
```

```python
order_data["Requested_Qty"] = order_data["مقدار اوليه درخواستي"] # مقدار/تعداد درخواستی
order_data["Allocated_Qty"] = order_data["مقدار تاييد شده"] # مقدار/تعداد تخصیص داده شده
order_data["Planner_code"] = order_data["پلنر"]
order_data["Stock_level"] = order_data["موجودي كل انبارها"]
order_data["Decision"] = order_data["ACTION"]  # APPROVE/REJECT و غیره
order_data["Delivered_Qty"] = order_data["مقدار تحويل شده"]

order_data["Allocation_Ratio"] = order_data["Allocated_Qty"]/order_data["Requested_Qty"].replace(0, 1)

#order_data["حواله در جریان"].fillna(0, inplace=True) #filling empty cells with 0
#order_data.count()

order_data["Date_billCreated"] = pd.to_datetime(order_data["تاريخ ايجاد حواله"], errors='coerce')
order_data["Date_plannerReceived"] = pd.to_datetime(order_data["تاریخ دريافت حواله توسط پلنر"], errors='coerce')

# چک تفاوت تخصیص و تحویل (برای دیدن تغییر توسط انبار)
order_data["Delivery_difference"] = order_data["Allocated_Qty"] - order_data["Delivered_Qty"] #منفی یعنی انبار کمتر داده 

# حذف ردیف‌های ناقص (برای ستون‌های کلیدی)
order_data= order_data.dropna(subset = ["Requested_Qty", "Allocated_Qty", "Stock_level"])

order_data.to_csv("cleaned_order_data.csv",index=False, encoding="utf-8")
print("دیتای تمیز ذخیره شد! تعداد ردیف‌های باقی‌مانده:", order_data.shape[0])

# نمونه ۵ ردیف تمیز شده
print("\nنمونه ۵ ردیف تمیز شده:")
print(order_data.head(5))
#print(order_data.dtypes)
#print(order_data.describe())
#print(order_data["میانگین مصرف کل"].mean())
#print(order_data.groupby(["شرح آیتم"])["Requested_Qty"].sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 

# آمار کلی نسبت تخصیص
print("میانگین نسبت تخصیص:", order_data["Allocation_Ratio"].mean())
print("حداقل و حداکثر نسبت تخصیص:", order_data["Allocation_Ratio"].min(), order_data["Allocation_Ratio"].max())

low_stock = order_data[order_data["Stock_level"] < 10]
print("تعداد رد در موجودی کم:", low_stock["Decision"].value_counts())

print("\nمیانگین تفاوت تخصیص و تحویل:", order_data["Delivery_difference"].mean())

# نمودار توزیع نسبت تخصیص
plt.figure(figsize=(8, 5))
sns.histplot(order_data["Allocation_Ratio"], bins=20, kde=True, color="blue")
plt.title("Allocatio_Ratio distribution")
plt.xlabel("Allocation_Ratio (0 to 1)")
plt.ylabel("Count")
plt.show()

# نمودار جعبه‌ای نسبت تخصیص بر اساس عملیات پلنر
plt.figure(figsize= (8, 5))
sns.boxplot(x="Decision", y="Allocation_Ratio", data=order_data, palette="Set2")
plt.title("Allocation_ratio vs. Decision")
plt.xlabel("Decision (APPROVE/REJECT)")
plt.ylabel("Allocation_Ratio")
plt.show()

# نمودار پراکندگی تخصیص و موجودی
plt.figure(figsize=(8, 5))
sns.scatterplot(x="Stock_level", y="Allocation_Ratio", data=order_data, hue="Decision", palette="deep")
plt.title("پراکندگی نسبت تخصیص و موجودی")
plt.xlabel("Stock_level")
plt.ylabel("Allocation_Ratio")
plt.show()

if order_data["Allocation_Ratio"].mean() < 0.5:
    print("پیشنهاد: نسبت تخصیص میانگین کمتر از 0.5 است. بررسی کنید چرا پلنرها کمتر تخصیص می‌دهند")
else:
    print("پیشنهاد: نسبت تخصیص خوب است، اما برای بهینه‌سازی، روی موارد رد تمرکز کنید")
```
